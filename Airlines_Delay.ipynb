```python
DelayF.dropna(subset = ["Cancelled"], inplace = True)
```

```python
DelayF["Month"] = DelayF["Month"].astype(int)
```

```python
from sklearn.model_selection import train_test_split

X= DelayF.drop("Cancelled", axis = 1)
y =  DelayF["Cancelled"].copy()

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import numpy as np

num_features = X_train.select_dtypes(include = [np.number]).columns

cat_features = DelayF.select_dtypes(exclude = [np.number]).columns
DelayF.drop(cat_features, axis = 1, inplace = True)
```

```python
num_pipeline = Pipeline([(
    ("imputer", SimpleImputer(strategy ='mean'))
)])

transformer = ColumnTransformer([
    ("num_pipeline", num_pipeline, num_features)
], remainder = "passthrough")
```

```python
X_train_transformed = transformer.fit_transform(X_train)
X_train_transformed
```

```python
columns_name = transformer.get_feature_names_out()
columns_name
```

```python
columns_name = transformer.get_feature_names_out()
```

```python
X_train_transformer_DelayF = pd.DataFrame(X_train_transformed, columns = columns_name)
X_train_transformer_DelayF
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier()

random_forest.fit(X_train_transformer_DelayF, y_train)
```

```python
random_forest.score(X_train_transformer_DelayF, y_train)
```

```python
random_forest.feature_importances_
```

```python
random_forest.feature_names_in_
```

```python
# prompt: help tme t visualize random_forest.feature_names_in_ and random_forest.feature_importances_
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Create a dataframe with feature names and importances
feature_importances_df = pd.DataFrame({
    "feature": random_forest.feature_names_in_,
    "importance": random_forest.feature_importances_
})

# Sort the dataframe by importance in descending order
feature_importances_df = feature_importances_df.sort_values(by="importance", ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x="importance", y="feature", data=feature_importances_df)
plt.title("Feature Importances")
plt.show()
```
