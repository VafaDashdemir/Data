```python
new_cols_dict = {
    "Timestamp": "time",
    "cnt" : "count",
    "t1": "temp_real_C",
    "t2": "tenp_feels_like_C",
    "hum": "humidity_percent",
    "wind_speed" : "wind_speed_kph",
    "weather_code": "weather",
    "is_holiday": "is_holiday",
    "is_weekend" : "is_weekend",
    "season": "season"

}



Bike.rename(new_cols_dict, axis = 1, inplace = True)
```

```python
Bike.humidity_percent = Bike.humidity_percent / 100
Bike.humidity_percent
```

```python
Bike.season.value_counts()
```

```python
season_dict = {
    "0.0": "spring",
    "1.0": "summer",
    "2.0": "autumn",
    "3.0": "winter"
}


weather_dict = {
    "1.0": "Clear",
    "2.0": "Scattered clouds",
    "3.0": "Broken clouds",
    "4.0": "Cloudy",
    "7.0": "Rain",
    "10.0": "Rain with thunderstormn",
    "26.0": "Snowfalling"

}

Bike.season = Bike.season.astype("str")

Bike.season = Bike.season.map(season_dict)

Bike.weather = Bike.weather.astype("str")

Bike.weather = Bike.weather.map(weather_dict)
```
