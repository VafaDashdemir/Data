```python
for i in range (1 ,100):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
def fact(x):
     total = 1
     for i in range(1, x+1):
         total*=i
     return total

print(fact(9))
```
