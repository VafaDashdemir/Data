```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Age', y='PerformanceRating', data=data)
plt.title('Age vs. Performance Rating')
plt.show()
```

```python
numerical_cols = ['Age', 'EmpHourlyRate', 'EmpEnvironmentSatisfaction', 'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany', 'PerformanceRating']
numerical_data = data[numerical_cols]

plt.figure(figsize=(12, 8))
corr_matrix = numerical_data.corr()

plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()

plt.xticks(np.arange(len(numerical_cols)), numerical_cols, rotation=45)
plt.yticks(np.arange(len(numerical_cols)), numerical_cols)

# Add annotations
for i in range(len(numerical_cols)):
    for j in range(len(numerical_cols)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='w')

plt.title('Correlation Heatmap')
plt.show()
```

```python
# Assuming numerical_cols is a list of numerical columns in your dataset
numerical_cols = ['Age', 'EmpHourlyRate', 'EmpEnvironmentSatisfaction', 'TotalWorkExperienceInYears', 'ExperienceYearsAtThisCompany', 'PerformanceRating']
numerical_data = data[numerical_cols]

# Add more advanced relationships using pair plot
sns.pairplot(data, vars=numerical_cols, hue='PerformanceRating', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()


```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Gender', 'EmpDepartment', and 'PerformanceRating' are columns in your dataset
gender_department_performance = data.groupby(['Gender', 'EmpDepartment'])['PerformanceRating'].mean().reset_index()

# Plotting with Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='EmpDepartment', y='PerformanceRating', hue='Gender', data=gender_department_performance, palette='pastel')
plt.title('Performance Rating by Gender and Department')
plt.xlabel('Department')
plt.ylabel('Mean Performance Rating')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python


# 1. Correlation Analysis
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix[['PerformanceRating']], annot=True, cmap='coolwarm')
plt.title('Correlation Matrix with PerformanceRating')
plt.show()
```

```python

# 2. Distribution and Descriptive Statistics
plt.figure(figsize=(10, 6))
sns.histplot(data['PerformanceRating'], kde=True)
plt.title('Distribution of PerformanceRating')
plt.show()
```
